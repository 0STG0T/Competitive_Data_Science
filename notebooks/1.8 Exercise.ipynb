# Apply + Function + One column

```python
def my_awesome_function(x):

    if x > 30:
        return 'more_30'
    else:
        return 'no_more_30'

rides_info['speed_avg_greater_30'] = rides_info['speed_avg'].apply(my_awesome_function) # Применяем функцию к столбцу

rides_info['speed_avg_greater_30'].value_counts() # Смотрим на результат
```

```python
rides_info['speed_avg_greater_30'].value_counts()
```

# Apply + lambda function

```python
my_awesome_function = lambda x: 'more_30' if x > 30 else 'no_more_30'

rides_info['speed_avg_greater_30'] = rides_info['speed_avg'].apply(my_awesome_function)

rides_info.head()
```

```python
rides_info['speed_avg_greater_30'].value_counts()
```

```python
two_cols = ['speed_avg', 'speed_max']

def A_devide_by_B(x):
    return x['speed_avg']/x['speed_max']

rides_info['new_feature'] = rides_info[two_cols].apply(A_devide_by_B, axis=1)

rides_info.head()
```

# Groupby + count / mean / sum / median 

```python
rides_info_gr = rides_info.groupby('car_id', as_index=False)[['rating', 'ride_duration', 'ride_cost']].mean()

rides_info_gr.head()
```

# Groupby + agg

```python
cols = ['rating', 'ride_duration', 'ride_cost']
rides_info_gr = rides_info.groupby('car_id', as_index=False)[cols].agg(['mean', 'sum']) # Производим группировку
rides_info_gr.columns = ['_'.join(c) for c in rides_info_gr.columns] # Убираем мульти индекс
rides_info_gr.head()
```

# Groupby + agg + tuples

```python
rides_info_gr = rides_info.groupby('car_id', as_index=False).agg(rating_mean = ('rating', 'mean'),
                                                                 ride_duration_sum = ('ride_duration', 'sum'),
                                                                 ride_cost_max = ('ride_cost', 'max'))


rides_info_gr.head(5)
```

# Groupby + transform = groupby + agg + merge

```python
# Делаем группировку
rides_info_gr = rides_info.groupby('car_id', as_index=False).agg(rating_mean = ('rating', 'mean')) 

# Приклеиваем полученные значения к исходной таблице
rides_info = rides_info.merge(rides_info_gr, on = 'car_id', how = 'left') 

# Смотрим результат
rides_info.head(5)
```

```python
rides_info['rating_mean'] = rides_info.groupby('car_id', as_index=False)['rating'].transform('mean')

# Смотрим результат
rides_info.head()
```

# Groupby + transform + function

```python
# Определяем нужную функцию
transform_function = lambda x: np.sum(x)/len(x)

# Применяем фнукцию для сгруппированных значение с сохранением индексов
rides_info['rating_mean'] = rides_info.groupby('car_id', as_index=False)['rating'].transform(transform_function)

# Смотрим результаты
rides_info.head()
```
